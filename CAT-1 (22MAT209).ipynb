## Q.1 a)

```python
n=21
rep=1000
xn_bar_val=[]
for i in range(0,rep+1):
    x=np.random.exponential(1,n)
    xn_bar=1/np.mean(x)
    xn_bar_val.append(xn_bar)
    
plt.hist(xn_bar_val,density=True,bins=40,color='blue')
z=np.linspace(-4,4,1000)
plt.xlim(0,3)
plt.plot(z,(stats.norm.pdf(z,1,1/np.sqrt(n))),c='r',linewidth=3)
```

```python
rep=1000
n_val=[20,50,100,200,500,1000]
for j in range(len(n_val)):
    n=n_val[j]
    xn_bar_val=[]
    for i in range(0,rep+1):
        x=np.random.exponential(1,n)
        xn_bar=1/np.mean(x)
        xn_bar_val.append(xn_bar) 
        
    plt.figure(figsize=(7,7))
    plt.subplot(2,3,j+1)    
    plt.hist(xn_bar_val,density=True,bins=40,color='blue')
    z=np.linspace(-4,4,1000)
    plt.xlim(0,2.5)
    plt.plot(z,(stats.norm.pdf(z,1,1/np.sqrt(n))),c='r',linewidth=3)
```

## Q.1 b)

```python
n=100
rep=1000
g = lambda x: x**4
g_val=[]
for i in range(0,rep+1):
    x=np.random.exponential(1,n)
    y=1/np.mean(x)
    g_val.append(g(y))
    
plt.hist(g_val,density=True,bins=40,color='blue')

new_mean=g(1)
new_var=16/n

z=np.linspace(-4,4,1000)
plt.xlim(0,3)
plt.plot(z,(stats.norm.pdf(z,new_mean,np.sqrt(new_var))),c='r',linewidth=3)
```

```python
rep=1000
g = lambda x: x**4
n_val=[20,50,100,200,500,1000]
for j in range(len(n_val)):
    n=n_val[j]
    g_bar_val=[]
    for i in range(0,rep+1):
        x=np.random.exponential(1,n)
        xn_bar=1/np.mean(x)
        g_bar_val.append(g(xn_bar))
        
    new_mean=g(1)
    new_var=16/n

    plt.figure(figsize=(7,7))
    plt.subplot(3,2,j+1)
    plt.hist(g_bar_val,density=True,bins=40,color='blue')
    z=np.linspace(-4,4,1000)
    plt.xlim(0,3)
    plt.plot(z,(stats.norm.pdf(z,new_mean,np.sqrt(new_var))),c='r',linewidth=3)
        
```
